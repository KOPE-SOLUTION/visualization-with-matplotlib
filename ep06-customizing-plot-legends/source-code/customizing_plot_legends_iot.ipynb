```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
hours = np.arange(0, 24)

greenhouse_a = 25 + 4 * np.sin(
    (hours - 7) * np.pi / 12
)
greenhouse_b = 24 + 3 * np.sin(
    (hours - 6) * np.pi / 12
)
outdoor = 27 + 6 * np.sin(
    (hours - 8) * np.pi / 12
)
```

```python
print("Hours:", hours.shape)
print("Greenhouse A:", greenhouse_a.shape)
print("Greenhouse B:", greenhouse_b.shape)
print("Outdoor:", outdoor.shape)
```

```python
fig, ax = plt.subplots(figsize=(9, 5))

ax.plot(
    hours,
    greenhouse_a,
    color="forestgreen",
    linewidth=2.5,
    label="Greenhouse A",
)
ax.plot(
    hours,
    greenhouse_b,
    color="royalblue",
    linewidth=2.5,
    linestyle="--",
    label="Greenhouse B",
)
ax.plot(
    hours,
    outdoor,
    color="darkorange",
    linewidth=2,
    linestyle=":",
    label="Outdoor sensor",
)

ax.set(
    title="Smart Farm Temperature Monitoring",
    xlabel="Hour",
    ylabel="Temperature (°C)",
    xlim=(0, 23),
)
ax.grid(alpha=0.25)

ax.legend(
    loc="upper left",
    title="Sensor location",
    ncols=1,
    frameon=True,
    fancybox=True,
    framealpha=1,
    borderpad=0.8,
    labelspacing=0.6,
    handlelength=2.5,
)

fig.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(9, 5))

ax.plot(hours, greenhouse_a, label="Greenhouse A")
ax.plot(hours, greenhouse_b, label="Greenhouse B")
ax.plot(hours, outdoor, label="Outdoor sensor")

ax.set(
    title="Temperature by Sensor Location",
    xlabel="Hour",
    ylabel="Temperature (°C)",
)
ax.grid(alpha=0.25)

ax.legend(
    loc="lower center",
    ncols=3,
    frameon=True,
)

fig.tight_layout()
plt.show()
```
